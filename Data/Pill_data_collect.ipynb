```python
#23000개  데이터 다 다운받기
import urllib.request
import pandas as pd
import re

# Load Dataset
data = pd.read_csv('./drug.csv')
data.head()

image_url = pd.DataFrame(data['큰제품이미지'][5318:])
image_url.head()

def downImage(img_url, img_name):
    result_url = './drug_image2/'
    urllib.request.urlretrieve(img_url, result_url+img_name+'.jpg')

for i, row in image_url.iterrows():
    print('row',i,'url', row['큰제품이미지'])
    if 'https://' in row['큰제품이미지']:
        downImage(row['큰제품이미지'], re.split('[/,-,<,>,|,(,). :]',data['품목명'][i])[0])
```

```python
#데이터 중 원형인 알약만 10000개 다운받기
import urllib.request
import pandas as pd
import re

data = pd.read_csv('./drug_file.csv')
data.head()

image_url = pd.DataFrame(data['큰제품이미지'])
image_url2 = pd.DataFrame(data['의약품제형'])
image_url.head()

def downImage(img_url, img_name):
    result_url = './drug_image3/'
    urllib.request.urlretrieve(img_url, result_url+img_name+'.jpg')

cnt=0
for i, row in image_url.iterrows():
    if cnt <10000:
        if 'https://' in row['큰제품이미지']:
            if  image_url2['의약품제형'][i] =='원형':
                cnt=cnt+1
                print(cnt,' row ',i,'url', row['큰제품이미지'], image_url2['의약품제형'][i])
                downImage(row['큰제품이미지'], re.split('[/,|,<,>,-,(,). :]',data['품목명'][i])[0])
    else:
        break
```
